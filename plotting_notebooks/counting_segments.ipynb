```python
from numpy import load
from sys import path

path.append("../")

from src.utils.misc import get_labels_counts
from src.utils.plots import plot_binary_labels
```

```python
def make_count_plot_info(
    path_to_data: str,
    dataset_name: str,
    output_folder: str,
    plot_title: str = "Distribution of Segments",
    labels_mapping: list[str] = ["negative", "positive"],
):
    data = load(path_to_data)
    labels_left = data["labels_left"]
    labels_right = data["labels_right"]
    # NOTE: I have checked and relaxation should be the negative label, and cognitive load the positive
    counts = get_labels_counts(
        labels_left, labels_right, labels_mapping
    )
    plot_binary_labels(
        counts=counts,
        title=plot_title,
        dataset_name=dataset_name,
        figsize=5,
        output_folder=output_folder,
    )
    print(counts)
```

# USILaughs

```python
make_count_plot_info(path_to_data="../data.nosync/usilaughs/segmented/EDA_segmented_noart.npz",
                     dataset_name="usilaughs",
                     output_folder="../final_visualizations/",
                     plot_title="Distribution of Segments, USILaughs",
                     labels_mapping=["Relaxation", "Cognitive Load"])
```
